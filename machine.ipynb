```python
import pandas as pd
import numpy as np

# Read the CSV file into a DataFrame
df = pd.read_csv('Pull_Data.csv')

# Access individual columns
rest_ranks = df['rest rank']
food_rank = df['Food rank']
count = 0
rate_sum = 0
food_rate_sum = 0
# Do something with the data
# print(column1, column2)
for rate in rest_ranks :
    count+=1
    rate_sum += rate
for rate in food_rank :
    food_rate_sum += rate

rate_avrage = rate_sum/count
food_avrage = food_rate_sum/count

print("number of rests : " ,count)
print("Avrage of ranks : " ,rate_avrage)

rank_array = rest_ranks.values
# Filter elements greater than avrage and save their indices
filtered_indices = [index for index, value in enumerate(rank_array) if value > rate_avrage]
filtered_array_by_rest_rank = [value for value in rank_array if value > rate_avrage]

food_array = food_rank.values
filtered_array_by_food_rank = [value for value in food_array if value > food_avrage]


print("Ranks array: " , rank_array)
# print(filtered_indices)
print("Array after filter by rate : " ,filtered_array_by_rest_rank)
print("Array after filter by food : " ,filtered_array_by_food_rank)



```
